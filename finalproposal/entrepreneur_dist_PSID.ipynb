### PSID family info

```python
data_path = "/Users/lingyi/Documents/PSID/PSID_data"
psid_data = {}

for year in range(1990, 2024):
    file_path = os.path.join(data_path, f"fam{year}.dta")
    if os.path.exists(file_path):
        psid_data[year] = pd.read_stata(file_path)
        print(f"✅ Loaded {file_path}")
```

```python
whether_map = {
    1990: 'V18703', 1991: 'V20003', 1992: 'V21303', 1993: 'V23162',
    1994: 'ER3096',  1995: 'ER6096',  1996: 'ER8193',  1997: 'ER11088',
    1999: 'ER13111', 2001: 'ER18489', 2003: 'ER21857', 2005: 'ER25838',
    2007: 'ER36856', 2009: 'ER42847', 2011: 'ER48169', 2013: 'ER53863',
    2015: 'ER60922', 2017: 'ER66974', 2019: 'ER72997', 2021: 'ER79074',
    2023: 'ER83049'
}
for year, var_name in whether_map.items():
    if year in psid_data:
        df = psid_data[year]
        if var_name in df.columns:
            df = df.rename(columns={var_name: 'whether_business'})
            psid_data[year] = df
            print(f"{year}: '{var_name}' → 'whether_business'  [Success]")
        else:
            print(f"{year}: vriable '{var_name}' no data！[Failed]")
    else:
        print(f"{year}: no data！")
```

```python

weight_map = {
    1990: 'V18943', 1991: 'V20243', 1992: 'V21547', 1993: 'V23361',
    1994: 'ER4160',  1995: 'ER7000',  1996: 'ER9251',  1997: 'ER12084',
    1999: 'ER16519', 2001: 'ER20394', 2003: 'ER24179', 2005: 'ER28078',
    2007: 'ER41069', 2009: 'ER47012', 2011: 'ER52436', 2013: 'ER58257',
    2015: 'ER65492', 2017: 'ER71570', 2019: 'ER77631', 2021: 'ER81958',
    2023: 'ER85812'
}

for year, var_name in weight_map.items():
    if year in psid_data:
        df = psid_data[year]
        if var_name in df.columns:
            df = df.rename(columns={var_name: 'fam_weight'})
            psid_data[year] = df
            print(f"{year}: '{var_name}' → 'fam_weight'  [Success]")
        else:
            print(f"{year}: vriable '{var_name}' no data！[Failed]")
    else:
        print(f"{year}: no data！")
```

```python
entre_weighted = {}
for year in weight_map:
    if year in psid_data and 'whether_business' in psid_data[year].columns and 'fam_weight' in psid_data[year].columns:
        df = psid_data[year]
        df_valid = df[df['fam_weight'] > 0].copy()
        
        if len(df_valid) > 0:
            # 
            yes_weighted = (df_valid['whether_business'] == 1) * df_valid['fam_weight']
            total_weight = df_valid['fam_weight']
            rate = yes_weighted.sum() / total_weight.sum()
            entre_weighted[year] = rate
        else:
            entre_weighted[year] = None
    else:
        entre_weighted[year] = None

rate_df = pd.Series(entre_weighted).reset_index()
rate_df.columns = ['year', 'entrepreneur_rate_weighted']
rate_df = rate_df.dropna()
print("\n=== entrepreneur share ===")
print(rate_df)
```

```python
plt.figure()
plt.plot(rate_df['year'], rate_df['entrepreneur_rate_weighted'], marker='o')
plt.grid(True)
plt.xlabel("Year")
plt.ylabel("Entrepreneur Rate")
plt.title("Entrepreneur Rate Over Time (PSID)")
plt.savefig('/Users/lingyi/Yi_PublicEcon/finalproposal/entrepreneursharepsid.eps', format='eps', dpi=1000 )
plt.show()
```
